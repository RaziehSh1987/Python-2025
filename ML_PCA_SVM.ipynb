```python
import numpy as np  
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv(".\diabetes.csv")
x=df.drop("Outcome", axis=1)
y=df["Outcome"]

x=np.array(x)
y=np.array(y)
```

```python
normalize=StandardScaler()
x=normalize.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x, y , test_size=0.2)

x_train.shape
```

```python
from sklearn import svm
model=svm.SVC(kernel='linear') #linear / rfb /sigmoid /..
model.fit(x_train,y_train)


y_pred_train=model.predict(x_train)
y_pred_test=model.predict(x_test)
print(y_pred_test)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score, confusion_matrix

acc_train=accuracy_score(y_true=y_train,y_pred=y_pred_train)

acc=accuracy_score(y_true=y_test,y_pred=y_pred_test)
# چند درصد درست پیش بینی کرده

rec=recall_score(y_true=y_test , y_pred=y_pred_test)
# از بین همه مثبت‌های واقعی، چندتاش رو درست پیدا کردیم؟

prec=precision_score(y_true=y_test , y_pred=y_pred_test)
# از بین اونهایی که گفته مریضیشون مثبت بوده  فقط چند درصدش واقعاً مثبت بوده

conf=confusion_matrix(y_true=y_test , y_pred=y_pred_test)

print(acc_train,"\n",acc,"\n", rec,"\n",prec , "\n", conf)
```

after PCA:

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=3)# I want to transform the 8 features (X datas) to 3 features
pca.fit(x_train)
x_new_train=pca.transform(x_train)
x_new_test=pca.transform(x_test)
```
